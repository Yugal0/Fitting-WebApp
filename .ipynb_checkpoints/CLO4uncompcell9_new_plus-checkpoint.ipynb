```python
@element(num_params=5, units=[" ", " ", " ", "",""],overwrite=True)
def ManjuIP(p,f):
    omega = 2 * np.pi * np.array(f)
    
    Rion, RCT, Rsol, Q, alpha = p[0], p[1], p[2], p[3], p[4]
    
    Z=[]
    for omg in omega:
        S=np.sqrt(((1j*omg)**alpha*Q+1/RCT)*-2*Rion)
        Zelem=mp.nstr(Rsol-2*Rion*mp.besselj(0,S)/(S*mp.besselj(1,S)),n=50)
        Zelem = Zelem.replace(' ','').replace('(','').replace(')','') 
        # print(Zelem)
        Z.append(Zelem)
        
    Z=np.asarray(Z,dtype=complex)
    # print(np.shape(Z)[1])
    # Z = np.shape(np.reshape(Z,np.shape(Z)[1]))
    # Z = np.shape(np.reshape(Z,np.shape(Z)[1]))

    return Z
```

```python
# f_high=200e3 #high
f_high=39897 #high
f_low=1.00016e-3 #Low 

def find_index_of_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

fhigh_idx=find_index_of_nearest(freq_list,f_high)
flow_idx=find_index_of_nearest(freq_list,f_low)

# print(fhigh_idx," ",flow_idx)

#crop the data, that falls in this frequency range
if(fhigh_idx<flow_idx):
    freq_to_fit=freq_list[fhigh_idx:flow_idx+1]
    zexp_to_fit=zexp[fhigh_idx:flow_idx+1]
else:
    freq_to_fit=freq_list[flow_idx:fhigh_idx+1]
    zexp_to_fit=zexp[flow_idx:fhigh_idx+1]
```

```python
# wt="unit"
wt="mod"
# wt="prop"
# wt="log_mod"
# wt="ModPlusUnit"
# 
def give_weights(z,f,wt):
    if (wt=="mod"):
        return np.concatenate((np.abs(z),np.abs(z)))
    elif (wt=="log_mod"):
        return np.concatenate((np.log(1+f)*np.real(z),np.log(1+f)*np.imag(z)))
    elif (wt=="prop"):
        return np.concatenate((np.real(z),np.imag(z)))
    elif (wt=="" or wt=="unit"):
        return np.ones(2*len(z))
    elif (wt=="ModPlusUnit"):
        return np.concatenate((np.real(z),np.imag(z)))

# print(np.shape(give_weights(zexp_to_fit,wt)))
# print(np.shape(zexp_to_fit))

# c_to_fit.fit(impedance=zexp_to_fit,frequencies=freq_to_fit,sigma=give_weights(zexp_to_fit,freq_to_fit,wt),bounds=((2,89,2e-6,0.95,2300,0,0,49,0,0),[2.2,91,2.6e-6,1,2500,np.inf,np.inf,51,np.inf,1]))
# c_to_fit.fit(impedance=zexp_to_fit,frequencies=freq_to_fit,sigma=give_weights(zexp_to_fit,freq_to_fit,wt),bounds=((2,89,2e-6,0.95,0,0,49,0,0),[2.2,91,2.6e-6,1,np.inf,np.inf,51,np.inf,1]))
# c_to_fit.fit(impedance=zexp_to_fit,frequencies=freq_to_fit,sigma=give_weights(zexp_to_fit,freq_to_fit,wt),bounds=((89,2e-6,0.95,0,0,2,0,0),[91,2.6e-6,1,np.inf,np.inf,3,np.inf,1]))
c_to_fit.fit(impedance=zexp_to_fit,frequencies=freq_to_fit,sigma=give_weights(zexp_to_fit,freq_to_fit,wt),bounds=((0,0,0,0,0,0,0,0),[np.inf,np.inf,1,np.inf,np.inf,np.inf,np.inf,1]))
```
